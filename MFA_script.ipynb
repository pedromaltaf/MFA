### __1__ Calculate the shear stress, shear rate and shear viscosity from the raw data in the files.

```python
# acrescentrar stress, strain e viscosity

radius = [10e-3, 25e-3]

for i,data in enumerate(dataframes):
    for df in data:
        radius_index = i % 2
        raio = radius[radius_index]
        df['Stress'] = (2*df['torque'])/(np.pi*(raio)**3)
        df['strain'] = (df['rot_speed']*(raio))/1e-4
        df['viscosity'] = df['Stress']/df['strain']
```

```python
def sem_calculator(dataframe):
    viscosities = pd.concat([df.select_dtypes(include=['number'])['viscosity'] for df in dataframe], axis=1)
    return viscosities.sem(axis=1).to_list()

sem_120 = sem_calculator(dataframes120)
sem_150 = sem_calculator(dataframes150)
sem_220 = sem_calculator(dataframes220)
sem_250 = sem_calculator(dataframes250)

sems = [sem_120, sem_150, sem_220, sem_250]
```

### __4__ Fit the experimental data to a suitable rheological model. Determine the model parameters and evaluate the goodness of fit.

```python
# Função para calcular MAE
def calculate_mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

# Função para calcular MSE
def calculate_mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Função para calcular RMSE
def calculate_rmse(y_true, y_pred):
    return np.sqrt(calculate_mse(y_true, y_pred))

# Função para calcular R²
def calculate_r2(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - (ss_res / ss_tot)

def top_3_models(best_metrics):
    sorted_metrics = {}
    for metric, model_dict in best_metrics.items():
        reverse_order = (metric == 'r2')
        sorted_models = sorted(model_dict.items(), key=lambda x: x[1], reverse=reverse_order)[:3]
        sorted_metrics[metric] = sorted_models
    return sorted_metrics

#---------------------------- metricas -----------------------------------------

# Define rheological models
def power_law_model(shear_rate, m, n):
    return m * shear_rate ** (n - 1)

def carreau_yasuda_model(shear_rate, eta_0, eta_inf, lambda_, a, n):
    return eta_inf + (eta_0 - eta_inf) * (1 + (lambda_ * shear_rate) ** a) ** ((n - 1) / a)

def bingham_model(shear_rate, tau_0, eta):
    return tau_0 / shear_rate + eta

def herschel_bulkley_model(shear_rate, tau_0, k, n):
    return (tau_0 + k * shear_rate ** n)/shear_rate

def cross_model(shear_rate, eta_0, eta_inf, lambda_, m):
    return eta_inf + (eta_0 - eta_inf) / (1 + (lambda_ * shear_rate) ** m)

def casson_model(shear_rate, k_0, k_1):
    return (k_0 + k_1*np.sqrt(shear_rate))**2 / shear_rate

def Ofoli_Morgan_Steffe_model(shear_rate, sigma_0, mu_inf, n1, n2):
    return ((sigma_0 / shear_rate) ** n1 + mu_inf * shear_rate ** (n2 - n1)) ** (1 / n1)
    
radius = [10e-3, 25e-3]
    
for i, df in enumerate(dataframes_tra):

    # Plotar os gráficos
    fig, ax = plt.subplots(figsize=(8, 6))


    print(samples[i])
    sample_name = samples[i].replace(" ", "")

    print('------------------------------------------------------',sep='\t')

    strains = np.linspace(min(df['strain']),max(df['strain']),60)
    
    n_mins = np.ones((len(strains), 4))
    n_maxs = np.ones((len(strains), 4))
    
    radius_index = i % 2
    raio = radius[radius_index]
    
    n_mins[:, i] = (2 * M0) / (np.pi * (raio)** 3 * strains)
    n_maxs[:, i] = (h**3 * rho * strains) / (12 * R * raio)

    # Plot dos limites máximo e mínimo
    ax.plot(strains, n_maxs[:, i],'-', label='$\eta_{sef}$', color='blue')
    ax.plot(strains, n_mins[:, i],'-', label='$\eta_{min}$', color='red')

    above_limits = (df['viscosity'] > n_mins[:, i]) & (df['viscosity'] > n_maxs[:, i])
    
    # Filtra os dados de viscosidade
    df_filtered = df[above_limits]

    # Verifica se a barra de erro também está dentro dos limites
    above_max = (df['viscosity'] - sems[i]) > n_mins[:, i]
    above_min = (df['viscosity'] - sems[i]) > n_maxs[:, i]
    
    # Aplica a filtragem com base nos erros
    valid_points = above_max & above_min

    # Aplica os filtros finais aos dados e sems
    df_filtered = df_filtered[valid_points]
    sem_filtered = [sem for sem, condition in zip(sems[i], valid_points) if condition]    

    # Dados experimentais
    ax.errorbar(df_filtered['strain'], df_filtered['viscosity'], yerr = sem_filtered, fmt='.',markersize=5, markerfacecolor='none', label="Experimental Data")

    shear_rate = df['strain']
    viscosity = df['viscosity']
    
    best_mae ={}
    best_mse ={}
    best_rmse ={}
    best_r2 ={}
    
    #-------------------------Tikhonov-------------------------
    
    # Calculate torque and solve shear stress
    torque = df['torque']
    rim_shear_rate = df['rot_speed'] * 2 * np.pi / 1e-4
    N_k = 550  # Number of discretization points
    shear_rate_ = np.linspace(0, rim_shear_rate.max(), N_k)
    d_gamma = shear_rate_[1] - shear_rate_[0]

    # Build coefficient matrix A
    A = np.zeros((len(rim_shear_rate), N_k))
    for j, gamma in enumerate(shear_rate_):
        for i, gamma_r in enumerate(rim_shear_rate):
            if gamma <= gamma_r:
                weight = 2 / 3 if j % 2 == 1 else 4 / 3 if j % 2 == 0 else 1 / 3
                A[i, j] = weight * 2 * np.pi * R**3 * gamma**2 * d_gamma / gamma_r**3

    # Create the tri-diagonal matrix for regularization
    beta = np.zeros((N_k, N_k))
    for k in range(1, N_k - 1):
        beta[k, k - 1] = 1
        beta[k, k] = -2
        beta[k, k + 1] = 1

    lambda_reg = 0.2
    regularized_matrix = A.T @ A + lambda_reg * beta.T @ beta

    # Solve for shear stress
    shear_stress = solve(regularized_matrix, A.T @ torque)
    torque_comp = A @ shear_stress
    
    shear_stress_comp = torque_comp * 2 / (np.pi * raio**3)

    # shear_strain
    # Calculate experimental viscosity
    viscosity_comp = shear_stress_comp/df['strain']

    ax.plot(df['strain'], viscosity_comp,'d',markersize=5, markerfacecolor='none', label='Tikhonov regularization')
    
    mae = calculate_mae(viscosity, viscosity_comp)
    mse = calculate_mse(viscosity, viscosity_comp)
    rmse = calculate_rmse(viscosity, viscosity_comp)
    r2 = calculate_r2(viscosity, viscosity_comp)
    
    best_mae['Tikhonov regularization'] = mae
    best_mse['Tikhonov regularization'] = mse
    best_rmse['Tikhonov regularization'] = rmse
    best_r2['Tikhonov regularization'] = r2
    
    print(f"Tikhonov regularization: MAE={mae:.5e}")
    print(f"Tikhonov regularization: MSE={mse:.5e}")
    print(f"Tikhonov regularization: RSME={rmse:.5e}")
    print(f"Tikhonov regularization: R2={r2:.5e}")    
#--------------------------------------------------------------------------------------------------  

    models = {
        "Power-Law": (power_law_model, [1, 1]),
        "Carreau-Yasuda": (carreau_yasuda_model, [1, 1, 1, 1, 1]),
        "Bingham": (bingham_model, [0.1, 1]),
        "Herschel-Bulkley": (herschel_bulkley_model, [0.1, 1, 1]),
        "Cross": (cross_model, [10, 0.1, 1, 1]),
        "Casson": (casson_model, [10, 1]),
        "OMS model": (Ofoli_Morgan_Steffe_model,[0.1,1,1,1]),

    }
    markers = {
        'Power-Law': 's',  # square
        'Carreau-Yasuda': 'o',  # circle
        'Bingham': '^',  # triangle up
        'Herschel-Bulkley': 'v',  # triangle down
        'Cross': 'x',  # cross
        'Casson': 'h',
        "OMS model":'p',
    }


    for name, (model_func, p0) in models.items():
        try:
            # Ajustar o modelo aos dados
            popt, _ = curve_fit(model_func, shear_rate, viscosity, p0=p0,maxfev=100000,ftol=1e-9,xtol=1e-9,gtol=1e-9)
            # Predição usando o modelo ajustado
            predicted = model_func(shear_rate, *popt)
            # Calcular MAE
            mae = calculate_mae(viscosity, predicted)
            mse = calculate_mse(viscosity, predicted)
            rmse = calculate_rmse(viscosity, predicted)
            r2 = calculate_r2(viscosity, predicted)

            # Plotar o modelo ajustado
            ax.plot(shear_rate, predicted, marker=markers[name],markersize=5, markerfacecolor='none', linestyle=' ', label=f"{name}")
            best_mae[name] = mae
            best_mse[name] = mse
            best_rmse[name] = rmse
            best_r2[name] = r2
            print(f"{name}: MAE={mae:.5e}")
            print(f"{name}: MSE={mse:.5e}")
            print(f"{name}: RSME={rmse:.5e}")
            print(f"{name}: R2={r2:.5e}")
            print(f"{name} parameters: {popt}")

        except RuntimeError:
            # Ignorar erros de ajuste
            print(f"{name} (fit failed)")
            
    # Exibir os top 3 modelos por métrica
    best_metrics = {'mae': best_mae, 'mse': best_mse, 'rmse': best_rmse, 'r2': best_r2}
    top_3_results = top_3_models(best_metrics)
    
    print()
    print('xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx', sep='\t')
    for metric, sorted_models in top_3_results.items():
        print(f"Top 3 Models based on {metric.upper()}:")
        for rank, (name, value) in enumerate(sorted_models, 1):
            print(f"  {rank}. {name}: {value:.5e}")

    print('xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx', sep='\t')
    
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel("Shear Rate (1/s)", fontsize=12)
    ax.set_ylabel("Viscosity (Pa·s)", fontsize=12)
    ax.set_title(sample_name, fontsize=15)
    ax.legend(loc="upper left", bbox_to_anchor=(1.05, 1))
    ax.grid(True)
    fig.savefig(f'{sample_name}_ex4.png', dpi=300,bbox_inches='tight')
    plt.show()
```

```python
radius = [10e-3, 25e-3]
    
for i, df in enumerate(dataframes_tra):

    # Plotar os gráficos
    fig, ax = plt.subplots(figsize=(8, 6))


    print(samples[i])
    sample_name = samples[i].replace(" ", "")
    print('------------------------------------------------------',sep='\t')

    strains = np.linspace(min(df['strain']),max(df['strain']),60)
    
    
    n_mins = np.ones((len(strains), 4))
    n_maxs = np.ones((len(strains), 4))
    
    radius_index = i % 2
    raio = radius[radius_index]
    
    n_mins[:, i] = (2 * M0) / (np.pi * (raio)** 3 * strains)
    n_maxs[:, i] = (h**3 * rho * strains) / (12 * R * raio)

    above_limits = (df['viscosity'] > n_mins[:, i]) & (df['viscosity'] > n_maxs[:, i])
    
    # Filtra os dados de viscosidade
    df_filtered = df[above_limits]

    # Verifica se a barra de erro também está dentro dos limites
    above_max = (df['viscosity'] - sems[i]) > n_mins[:, i]
    above_min = (df['viscosity'] - sems[i]) > n_maxs[:, i]
    
    # Aplica a filtragem com base nos erros
    valid_points = above_max & above_min

    # Aplica os filtros finais aos dados e sems
    df_filtered = df_filtered[valid_points]
    sem_filtered = [sem for sem, condition in zip(sems[i], valid_points) if condition]    

    # Dados experimentais
    ax.errorbar(df_filtered['strain'], df_filtered['viscosity'], yerr = sem_filtered, fmt='.',elinewidth = .7, markersize=10, label="Experimental Data")

    shear_rate = df['strain']
    viscosity = df['viscosity']
    
    #-------------------------Tikhonov-------------------------
    
    # Calculate torque and solve shear stress
    torque = df['torque']
    rim_shear_rate = df['rot_speed'] * 2 * np.pi / 1e-4
    N_k = 550  # Number of discretization points
    shear_rate_ = np.linspace(0, rim_shear_rate.max(), N_k)
    d_gamma = shear_rate_[1] - shear_rate_[0]

    # Build coefficient matrix A
    A = np.zeros((len(rim_shear_rate), N_k))
    for j, gamma in enumerate(shear_rate_):
        for i, gamma_r in enumerate(rim_shear_rate):
            if gamma <= gamma_r:
                weight = 2 / 3 if j % 2 == 1 else 4 / 3 if j % 2 == 0 else 1 / 3
                A[i, j] = weight * 2 * np.pi * R**3 * gamma**2 * d_gamma / gamma_r**3

    # Create the tri-diagonal matrix for regularization
    beta = np.zeros((N_k, N_k))
    for k in range(1, N_k - 1):
        beta[k, k - 1] = 1
        beta[k, k] = -2
        beta[k, k + 1] = 1

    lambda_reg = 0.2
    regularized_matrix = A.T @ A + lambda_reg * beta.T @ beta

    # Solve for shear stress
    shear_stress = solve(regularized_matrix, A.T @ torque)
    torque_comp = A @ shear_stress
    
    shear_stress_comp = torque_comp * 2 / (np.pi * raio**3)

    # shear_strain
    # Calculate experimental viscosity
    viscosity_comp = shear_stress_comp/df['strain']

    ax.plot(df['strain'], viscosity_comp,'d',markersize=5, markerfacecolor='none', label='Tikhonov regularization')
    
    
#--------------------------------------------------------------------------------------------------  

    models = {
        "Power-Law": (power_law_model, [1, 1]),
        "Carreau-Yasuda": (carreau_yasuda_model, [1, 1, 1, 1, 1]),
        "Bingham": (bingham_model, [0.1, 1]),
        "Herschel-Bulkley": (herschel_bulkley_model, [0.1, 1, 1]),
        "Cross": (cross_model, [10, 0.1, 1, 1]),
        "Casson": (casson_model, [10, 1]),
        "OMS model": (Ofoli_Morgan_Steffe_model,[0.1,1,1,1]),
    }
    
    markers = {
        'Power-Law': 's',  # square
        'Carreau-Yasuda': 'o',  # circle
        'Bingham': '^',  # triangle up
        'Herschel-Bulkley': 'v',  # triangle down
        'Cross': 'x',  # cross
        'Casson': 'h',
        "OMS model":'p',
    }


    for name, (model_func, p0) in models.items():
        try:
            # Ajustar o modelo aos dados
            popt, _ = curve_fit(model_func, shear_rate, viscosity, p0=p0,maxfev=100000,ftol=1e-9,xtol=1e-9,gtol=1e-9)

            # Predição usando o modelo ajustado
            predicted = model_func(shear_rate, *popt)

            # Plotar o modelo ajustado
            ax.plot(shear_rate, predicted, marker=markers[name],markersize=5, markerfacecolor='none', linestyle=' ', label=f"{name}")

        except RuntimeError:
            # Ignorar erros de ajuste
            print(f"{name} (fit failed)")
            
    # Exibir os top 3 modelos por métrica
    best_metrics = {'mae': best_mae, 'mse': best_mse, 'rmse': best_rmse, 'r2': best_r2}
    top_3_results = top_3_models(best_metrics)
    

    if raio == 0.025:
        ax.set_xlim([1e3,1e5])
        ax.set_ylim([1e-3,1e2])
        
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel("Shear Rate (1/s)", fontsize=12)
    ax.set_ylabel("Viscosity (Pa·s)", fontsize=12)
  
    ax.set_title(sample_name, fontsize=15)
    ax.legend(loc="upper left", bbox_to_anchor=(1.05, 1))
    ax.grid(True)
    fig.savefig(f'{sample_name}_ex4a.png', dpi=300,bbox_inches='tight')
    plt.show()
```
